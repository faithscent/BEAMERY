# Data Merging

When merging the data the use of the 'pd.merge' which acts like the SQL Join function. It will require the following:

    a. Unique Identifier
    b. The DataFrames to be merged
    c. The type of join to be used (Inner)
    
For the contact and dim contact dataframe the merging will occur on an outer, this will ensure that we have all the data samples from all three dataframes. 

```python
#Merge the first two, beamery_contact and beamery_dim_contact
beamery_merge = pd.merge(beamery_contact, beamery_dim_contact, on='contact_id')
beamery_merge
```

```python
#Merge beamery_vac_stage with beamery_merge
#The contact_id on the left table and candidate_id on the right since the candidate id is mapped to the contact_id
beamery_merge = pd.merge(beamery_merge, beamery_vac_stage,left_on='contact_id',right_on='candidate_id')
beamery_merge
```

To calculate how many candidates have passed screening: 
    
    They will have both a screen date and and an interview date

```python
#Loop through dataframe row entered_screen_date and entered_interview_date when a row has both append to the new array 

passed_screening = [
    beamery_merge.iloc[i]
    for i in range(len(beamery_merge))
    if beamery_merge['entered_screen_date'][i] != 0
    and beamery_merge['entered_interview_date'][i] != 0
]
```

```python
#Create a pie chart of how many passed compared to the total 
labels = 'Passed', 'Failed'
sizes = [len(passed_screening), len(beamery_merge) - len(passed_screening)]
explode = (0, 0.1)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
